# Usage Example

```python
from metaheuropt.solvers import GA, PSO, MVO, CMAES, GWO, JADE, NSGAII, WOA
from metaheuropt.core import Optimizer
from metaheuropt.analysis import ResultsAnalyzer
import numpy as np
```

```python
def sphere(x):
    return np.sum(np.square(x))
```

```python
dims = 10
bounds = (np.array([-5.12] * dims), np.array([5.12] * dims))
num_runs = 5
max_iter = 100

solvers = [
    GA(bounds=bounds, pop_size=50, max_iter=max_iter),
    PSO(bounds=bounds, pop_size=50, max_iter=max_iter),
    MVO(bounds=bounds, pop_size=50, max_iter=max_iter),
    CMAES(bounds=bounds, pop_size=50, max_iter=max_iter),
    GWO(bounds=bounds, pop_size=50, max_iter=max_iter),
    JADE(bounds=bounds, pop_size=50, max_iter=max_iter),
    NSGAII(bounds=bounds, pop_size=50, max_iter=max_iter),
    WOA(bounds=bounds, pop_size=50, max_iter=max_iter),
]
```

```python
opt = Optimizer(solvers=solvers, obj_func=sphere, num_runs=num_runs)
opt.run(save_results=True, results_folder="demo_results")
```

```python
# Show the best fitness found across all runs for each solver
for solver_name in opt.results.keys():
    best_data = opt.get_overall_best(solver_name)
    print(f"Algorithm: {solver_name}")
    print(f"Best Fitness Found: {best_data['fitness']:.2e}")
    print(f"Best Solution (first 3 dims): {best_data['solution'][:3]}\n")
```

```python
analyzer = ResultsAnalyzer("demo_results")

analyzer.plot_convergence()
analyzer.plot_boxplots()
analyzer.plot_radar()
```
